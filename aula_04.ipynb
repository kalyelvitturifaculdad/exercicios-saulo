## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

## Exemplo Resolvido 2: Calculadora Simples

**O problema:** Peça dois números e um operador matemático (`+`, `-`, `*` ou `/`) ao usuário. Imprima o resultado da operação.

```python
num1 = float(input("Digite o primeiro número: "))
op = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

if op == "+":
    print(f"Resultado: {num1 + num2}")
elif op == "-":
    print(f"Resultado: {num1 - num2}")
elif op == "*":
    print(f"Resultado: {num1 * num2}")
elif op == "/":
    # Cuidados extras em programação: não se divide por zero!
    if num2 == 0:
        print("Erro: Divisão por zero.")
    else:
        print(f"Resultado: {num1 / num2}")
else:
    print("Operador inválido!")
```

### Desafio 2: Aumento de Salário
Faça um programa que calcule o aumento de salário de um funcionário.
- Para salários maiores que R$ 1500,00, aumento de 10%.
- Para salários menores ou iguais a R$ 1500,00, aumento de 15%.
Multiplicar um valor por 1.10 é dar 10% de aumento.

```python
# Seu código aqui

salario = float(input("Qual o salário atual? "))
if salario >= 1500:
  aumento = salario * 1.10
  print("O novo salário é: ", aumento)
  print("O percentual de aumento foi de: 10%")
elif salario <= 1500:
  aumento = salario * 1.15
  print("O novo salário é: ", aumento)
  print("O percentual de aumento foi de : 15%")

```

- Isósceles: 2 lados iguais;
